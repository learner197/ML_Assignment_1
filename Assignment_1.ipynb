```python
# x = input data
# y = output data
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#input of training data
x_train=pd.read_csv("a1-x-train.txt",names=["column"])
# output of training data
y_train=pd.read_csv("a1-y-train.txt",names=["output"])
```

```python
x_train
```

```python
len(x_train),len(y_train)
```

```python
y_train
```

```python

input_train=[]
for temp in range(len(x_train)):
    col_1=(float(x_train["column"][temp].split(" ")[0]))
    col_2=(float(x_train["column"][temp].split(" ")[1]))
    input_train.append([col_1,col_2])
input_train
```

```python
output_train=[]
for temp in range(len(y_train)):
    output_train.append(y_train["output"][temp])
output_train
```

```python
print("input features of 0th row:",input_train[0])
print("output value of 0th row:",output_train[0])
```

```python
len(output_train)
```

# Creating a dataframe

```python
df=pd.DataFrame(input_train,columns=["col_1","col_2"])
df
```

```python
# final dataframe
df["values"]=output_train
df
```

# Scatter plots

```python
# how output varies with col_1
plt.scatter(df["col_1"],df["values"])
plt.xlabel('col_1 values')
plt.ylabel('output values')
plt.title('how actual output varies with col_1')
```

```python
# how output varies with col_2
plt.scatter(df["col_2"],df["values"])
plt.xlabel('col_2 values')
plt.ylabel('output values')
plt.title('how actual output varies with col_2')
```

# Predicting on training data

```python
#importing linear regression model from scikit learn library
from sklearn.linear_model import LinearRegression
algo=LinearRegression()

algo.fit(input_train,output_train)
```

```python
# to get slope of linear eq. use .coef_
m=algo.coef_
m
```

```python
# to get intercept use .intercept_
c=algo.intercept_
c
```

```python
output_train_predicted=[]
for temp in range(len(input_train)):
    pred_val=m[0]*input_train[temp][0]+m[1]*input_train[temp][1]+c
    output_train_predicted.append([pred_val])
output_train_predicted
```

```python
# predicted output vs actual output on col_1
plt.scatter(df["col_1"],df["values"])
plt.xlabel('col_1 values')
plt.ylabel('actual output values')
plt.plot(df["col_1"],output_train_predicted,colDor='r')
plt.legend(['predicted output values'])
plt.title('predicted output vs actual output on col_1')
```

```python
# predicted output vs actual output on col_2
plt.scatter(df["col_2"],df["values"])
plt.xlabel('col_2 values')
plt.ylabel('actual output values')
plt.plot(df["col_2"],output_train_predicted,color='r')
plt.legend(['predicted output values'])
plt.title('predicted output vs actual output on col_2')
```

```python
# how predicted output varies with col_1
plt.scatter(df["col_1"],output_train_predicted)
plt.xlabel('col_1 values')
plt.ylabel('predicted output values')
plt.title('how predicted output varies with col_1')
```

```python
# how predicted output varies with col_2
plt.scatter(df["col_2"],output_train_predicted)
plt.xlabel('col_2 values')
plt.ylabel('predicted output values')
plt.title('how predicted output varies with col_2')
```

```python
# how actual and predicted output varies
plt.scatter(output_train_predicted,output_train)
plt.xlabel('actual output')
plt.ylabel('predicted output')
plt.title('how actual and predicted output varies')
```

# Training Accuracy

```python
# Model Accuracy on training data
from sklearn.metrics import r2_score
train_data_accuracy=r2_score(output_train_predicted,output_train)
train_data_accuracy
```
